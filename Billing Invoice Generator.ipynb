```python
no_of_items = int(input("Enter the number of items: "))

items_dict = {}

for i in range(no_of_items):
    item,quantity,cost = input("Enter Item name, item quantity and cost of each item seperated by space: ").split(" ")
    items_dict[item] = [int(quantity),int(cost)] #create a dictionary of items

#used to find the longest word in items for proper spacing    
max_word_len = max([len(i) for i in items_dict.keys()])

#for proper spacing acc. to header. i.e Item name\t is equivalent to 15 characters
if max_word_len < 15:
    max_word_len = 15

#used to find the longest number in quantity for proper spacing 
max_quan_len = max([len(str(items_dict[j][0])) for j in items_dict.keys()])

#for proper spacing acc. to header. i.e Quantity\t is equivalent to 15 characters
if max_quan_len < 15:
    max_quan_len = 15
print("\n")
print("*"*50)
print("\n\t\tDesai Mart\n\t\tBorivali - W\n\t\tMumbai-92\n")
    
print("="*50)    
    
print("\tItem name\tQuantity\tCost")
print("="*50)

sum = 0
#:<{len} used to provide spaces
for i in items_dict.keys():
    print(f"\t{i:<{max_word_len}} {items_dict[i][0]:<{max_quan_len}} {(items_dict[i][1])*(items_dict[i][0])}$")
    sum += (items_dict[i][1])*(items_dict[i][0])

print("="*50) 
# print("\n")
print("\t  " + " " * max_word_len + " " * max_quan_len + "Total")
print("\t  " + " " * max_word_len + " " * max_quan_len + str(sum) + "$")
# print("\n")
print("="*50)   

print("\n\tThank You for shopping with us!\n")
print("*"*50) 
```
